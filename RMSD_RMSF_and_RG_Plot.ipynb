```python
import numpy as np
import matplotlib.pyplot as plt

# Load RMSD data from the first XVG file
data1 = np.loadtxt('QK_rmsd.xvg', comments=['@', '#'])

# Load RMSD data from the second XVG file
data2 = np.loadtxt('seq3_rmsd.xvg', comments=['@', '#'])

# Convert time from ps to ns (assuming time is in the first column)
time_ns = data1[:, 0] / 1000  # Convert ps to ns

# Extract RMSD values (assuming RMSD is in the second column)
rmsd1 = data1[:, 1]
rmsd2 = data2[:, 1]

# Set up font properties for the plot
font = {'family': 'serif', 'size': 14}  # Use serif font and set size
plt.rc('font', **font)
plt.rcParams.update({'figure.autolayout': True})  # Avoid layout issues

# Plot RMSD data
plt.figure(figsize=(10, 6))

plt.plot(time_ns, rmsd1, label='QK Peptide (Positive Control)', color='blue', linewidth=2)
plt.plot(time_ns, rmsd2, label='VMP3', color='red', linewidth=2)

# Customize labels and title with appropriate font sizes
plt.xlabel('Time (ns)', fontsize=16, weight='bold')
plt.ylabel('RMSD (Å)', fontsize=16, weight='bold')
plt.title('Root Mean Square Deviation (RMSD) vs. Time', fontsize=18, weight='bold')

# Customize legend
plt.legend(fontsize=14, loc='upper right', frameon=False)

# Customize ticks on both axes
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add gridlines for clarity
plt.grid(True, linestyle='--', linewidth=0.5)

# Calculate average difference between the two datasets
average_difference = np.mean(np.abs(rmsd1 - rmsd2))
print("Average Difference:", average_difference, "Å")

# Save the plot as high-quality JPEG and PDF formats
plt.savefig('rmsd_plot_high_quality.jpg', format='jpg', dpi=300, bbox_inches='tight')  # High-resolution JPEG
plt.savefig('rmsd_plot_high_quality.pdf', format='pdf', dpi=300, bbox_inches='tight')  # High-resolution PDF

# Display the plot
plt.show()
```

```python
# Plot RMSD (for 3 XVG files)
import numpy as np
import matplotlib.pyplot as plt

# Load RMSD data from the XVG files
data1 = np.loadtxt('rmsd_complex_B7_IL1.xvg', comments=['@', '#'])
data2 = np.loadtxt('rmsd_complex_positive_IL1.xvg', comments=['@', '#'])
data3 = np.loadtxt('rmsd_complex_other.xvg', comments=['@', '#'])

# Convert time from ps to ns (assuming time is in the first column)
time_ns = data1[:, 0] / 1000  # Convert ps to ns

# Extract RMSD values (assuming RMSD is in the second column)
rmsd1 = data1[:, 1]
rmsd2 = data2[:, 1]
rmsd3 = data3[:, 1]

# Plot RMSD data
plt.figure(figsize=(10, 6))
plt.plot(time_ns, rmsd1, label='RMSD File 1', color='blue')
plt.plot(time_ns, rmsd2, label='RMSD File 2', color='red')
plt.plot(time_ns, rmsd3, label='RMSD File 3', color='green')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (Å)')
plt.title('Root Mean Square Deviation (RMSD) vs. Time')
plt.legend()
plt.grid(True)

# Annotate each file outside the plot box with smaller font size
plt.text(time_ns[-1], rmsd1[-1], 'B7-IL1R1 Complex', fontsize=8, color='blue', ha='right', va='bottom')
plt.text(time_ns[-1], rmsd2[-1], 'Positive Control Complex', fontsize=8, color='red', ha='right', va='bottom')
plt.text(time_ns[-1], rmsd3[-1], 'Other Complex', fontsize=8, color='green', ha='right', va='bottom')

# Calculate average difference between the datasets
average_difference_1_2 = np.mean(np.abs(rmsd1 - rmsd2))
average_difference_1_3 = np.mean(np.abs(rmsd1 - rmsd3))
average_difference_2_3 = np.mean(np.abs(rmsd2 - rmsd3))

print("Average Difference 1-2:", average_difference_1_2, "Å")
print("Average Difference 1-3:", average_difference_1_3, "Å")
print("Average Difference 2-3:", average_difference_2_3, "Å")

# Save the plot as JPEG and PDF
plt.savefig('rmsd_plot.jpg', bbox_inches='tight')  # JPEG format
plt.savefig('rmsd_plot.pdf', bbox_inches='tight')  # PDF format

plt.show()
```

```python
# RMSD Of Error in VEGF

import numpy as np
import matplotlib.pyplot as plt

# Load RMSF data from the first XVG file
data1 = np.loadtxt('QK_rmsf-protein.xvg', comments=['@', '#'])

# Load RMSF data from the second XVG file
data2 = np.loadtxt('Seq3_rmsf.xvg', comments=['@', '#'])

# Extract residue index (assuming residue index is in the first column)
residue_index1 = data1[:, 0]
residue_index2 = data2[:, 0]

# Extract RMSF values (assuming RMSF is in the second column)
rmsf1 = data1[:, 1]
rmsf2 = data2[:, 1]

# Find the minimum length between the two datasets
min_length = min(len(residue_index1), len(residue_index2), len(rmsf1), len(rmsf2))

# Trim all arrays to the same length
residue_index_trimmed = residue_index1[:min_length]
rmsf1_trimmed = rmsf1[:min_length]
rmsf2_trimmed = rmsf2[:min_length]

# Plot RMSF data
plt.figure(figsize=(10, 6))
plt.plot(residue_index_trimmed, rmsf1_trimmed, label='B7-IL1R1 Complex', color='blue', linewidth=2)
plt.plot(residue_index_trimmed, rmsf2_trimmed, label='Positive Control Complex', color='red', linewidth=2)
plt.xlabel('Residue Index')
plt.ylabel('RMSF (Å)')
plt.title('Root Mean Square Fluctuation (RMSF) per Residue')
plt.legend()
plt.grid(True)

# Save the plot as high-resolution images for journals
plt.savefig('rmsf_plot.jpg', bbox_inches='tight', dpi=300)  # JPEG format
plt.savefig('rmsf_plot.pdf', bbox_inches='tight', dpi=300)  # PDF format

plt.show()
```
