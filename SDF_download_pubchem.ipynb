```python
import requests
import zipfile
import glob
def download_sdf(pubchem_id):
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{pubchem_id}/record/SDF/?record_type=3d"
    response = requests.get(url)
    
    if response.status_code == 200:
        sdf_data = response.text
        file_name = f"pubchem_{pubchem_id}.sdf"
        
        with open(file_name, 'w') as file:
            file.write(sdf_data)
        
        print(f"SDF file downloaded successfully: {file_name}")
    else:
        print(f"Error downloading SDF file. PubChem ID: {pubchem_id}, HTTP Status Code: {response.status_code}")

# Read PubChem IDs from a text file
filename = "ML_NA550_Pubchem.txt"  # Replace with your text file name

with open(filename, 'r') as file:
    pubchem_ids = file.read().splitlines()

# Download SDF files for each PubChem ID
for pubchem_id in pubchem_ids:
    download_sdf(int(pubchem_id))  # Ensure PubChem ID is converted to an integer


```

```python

with zipfile.ZipFile("output.zip", "w") as zf:
    zf.write("*.sdf")
```
